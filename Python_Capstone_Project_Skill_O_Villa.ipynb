## Task 2
Compute the ***missing values*** by ***running average*** in column **Cost** using ***"For" loop***

```python
# Load Project Data
project = pd.read_csv("Project.csv")                  # pd.read_csv("file_name.extension") 
```

```python
# Computing missing values by running average in column Cost using "For" loop

# initialize running Total and count variable for itteration
running_total = 0
count = 0
```

```python
# Loop thorugh each row to fill missing cost 
for avg_cost in range(len(project)):
    if pd.isnull(project.loc[avg_cost, "Cost"]):
        running_avg = running_total / count if count > 0 else 0
        project.loc[avg_cost, "Cost"] = running_avg
    else:
        running_total += project.loc[avg_cost, "Cost"]
        count += 1
```

```python
# Save the updated dataframe
project.to_csv("Project.csv", index = False)
```

```python
# Output: Task 2
project
```

## Task 5
Generate a New column called ***Bonus***. and give a ***5%*** bonus to ***finished*** projects. in the ***final*** DataFrame

```python
# Lets Load the data from the final version dataframe final.csv file
final = pd.read_csv("Final.csv")                        # final.csv is loaded as we have created a final version after merging
```

```python
# Adding a bonus column with a anonymous function for 5% bonus with cost column
final["Bonus"] = final.apply(lambda row: round(row["Cost"] * 0.05, 2) if row["Status"] == "Finished" else 0, axis= 1)
```

```python
# Save the final DataFrame into Final.csv
final.to_csv("Final.csv", index = False)
```
